```python
import time
import matplotlib.pyplot as plt
from dynamic_array_seq import Dynamic_Array_Seq
```

```python
arr = Dynamic_Array_Seq()
times = []

for i in range(1, 10001):
    start = time.perf_counter()
    arr.insert_last(i)
    end = time.perf_counter()
    times.append((end - start) * 1e6)  # microseconds
```

```python
plt.figure(figsize=(12, 5))
plt.plot(times, label="Insert time (Âµs)")
plt.xlabel("Operation index")
plt.ylabel("Time per insert (microseconds)")
plt.title("Empirical Time per insert_last() (should be O(1) amortized)")
plt.legend()
plt.grid(True)
plt.show()
# The above code tests the performance of the insert_last method in Dynamic_Array_Seq
# by measuring the time taken for each insertion and plotting it.
```

```python
import numpy as np
cumulative_time = np.cumsum(times)
plt.plot(cumulative_time)
plt.xlabel("Number of insertions")
plt.ylabel("Cumulative time (Âµs)")
plt.title("Cumulative Time vs. Insertions")
plt.grid(True)
plt.show()
```
